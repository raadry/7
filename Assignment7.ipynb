```python


print("1. Get a translation")
print("2. Add a translation")
selection = input("Enter a number:")

while selection  !='0':

    if selection == '1':
        phrase = input("Enter word or phrase: ")

        match = False
        german = open('german.txt','r')
        line = german.readline().rstrip()

        while line!='':
            translation = line
            gtrans = german.readline().rstrip()

            if phrase == translation:
                print("German: ", gtrans)
                match = True
            line = german.readline().rstrip()

        if match == False:
            print("No translation found.")
        german.close()


        match = False
        french = open('french.txt','r')
        line = french.readline().rstrip()

        while line!='':
            translation = line
            ftrans = french.readline().rstrip()

            if phrase == translation:
                print("French: ", ftrans)
                match = True
            line = french.readline().rstrip()

        french.close()



        match = False
        spanish = open('spanish.txt','r')
        line = spanish.readline().rstrip()

        while line!='':
            translation = line
            strans = spanish.readline().rstrip()

            if phrase == translation:
                print("Spanish: ", strans)
                match = True
            line = spanish.readline().rstrip()


        spanish.close()
        
        print("1. Get a translation")
        print("2. Add a translation")
        print("0. Exit")

        selection = input("What would you like to do: ")
        
        while selection not in ['1','2','0']:
            input("Invalid response. Try again: ")








    elif selection == '2':

        german = open('german.txt','a')
        french = open('french.txt','a')
        spanish = open('spanish.txt','a')

        newphrase = input("Enter word or phrase to translate: ")
        gnewphrase = input("Enter translation in German: ")
        fnewphrase = input("Enter translation in French: ")
        snewphrase = input("Enter translation in Spanish: ")

        german.write(newphrase+"\n")
        german.write(gnewphrase+"\n")

        french.write(newphrase+"\n")
        french.write(fnewphrase+"\n") 

        spanish.write(newphrase+"\n")
        spanish.write(snewphrase+"\n")
        
        print("1. Get a translation")
        print("2. Add a translation")
        print("0. Exit")

        selection = input("What would you like to do: ")
        
        while selection not in ['1','2','0']:
            input("Invalid response. Try again: ")
```
